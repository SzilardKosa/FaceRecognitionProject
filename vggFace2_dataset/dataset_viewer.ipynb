We use this notebook to display elements from the dataset.

```python
# This library let's us work with hdf5 format files.
import tables
# Fundamental package for scientific computing with Python.
import numpy as np
# Used for image processing and face detection.
import cv2
# Activating the notebook style display.
%matplotlib inline
import matplotlib.pyplot as plt

# Opening the dataset.
hdf5_path = 'dataset.hdf5'
hdf5_file = tables.open_file(hdf5_path, mode='r')

data_num = hdf5_file.root.images.shape[0]
plt.rcParams['figure.figsize']=[16,16]
  
# We display the images in tight tables (default: 20x20 picture)
# Note: The notebook throws an error if there aren't enough picture to fill a table,
# but the pictures will be displayed nevertheless
for j in range(1):
    fig, ax_array = plt.subplots(20,20)
    axes = ax_array.flatten()
    for i, ax in enumerate(axes):
        ax.imshow(hdf5_file.root.images[i+j*400]*255, cmap='gray_r')
    plt.setp(axes, xticks=[], yticks=[], frame_on=False)
    plt.tight_layout(h_pad=0, w_pad=0)
        
hdf5_file.close()
```
